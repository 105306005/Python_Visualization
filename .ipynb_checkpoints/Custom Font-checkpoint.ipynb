# Customize Font with Matplotlib

Import custom fonts into your Python data visualizations. 

These techniques can also be used with other data visualization packages such as Seaborn. 

## Step1: Import some Libraries

```python
import matplotlib.font_manager
import matplotlib.pyplot as plt
%matplotlib inline

# for displaying list of all available fonts
from IPython.core.display import HTML

# for Google Fonts
import os
from matplotlib import font_manager as fm, rcParams
```

## Step2: View Available fonts to Matplotlib

see more information [here](http://jonathansoma.com/lede/data-studio/matplotlib/list-all-fonts-available-in-matplotlib-plus-samples/)

```python
def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

## Step3: Start use those fonts

```python
# Create a blank figure
fig, ax = plt.subplots()

# choose a Custom font 
custom_font1 = "Phosphate"
custom_font2 = "Brush Script MT"

# set title, lables, text with the font
ax.set_title("Custom font title", fontname=custom_font1, size=20)
ax.set_xlabel("X label", fontname=custom_font1, size=20)
ax.set_ylabel("Y label", fontname=custom_font1, size=20)
plt.text(0.2, 0.5, "random text", fontname=custom_font2, size=20)

# set axis tick labels with the font
for tick in ax.get_xticklabels():
    tick.set_fontname(custom_font2)
    
for tick in ax.get_yticklabels():
    tick.set_fontname(custom_font2)
```

## Advance1: use downloaded font (such as Google Fonts)

```python
# Create a blank figure
fig, ax = plt.subplots()


fpath = '/Users/maggiesun/Documents/NYU課業/NYU一下/learning/Machine_Learning_and_Data_Science/Python-Visualization/Custom_Font/Lato-Regular.ttf'
prop = fm.FontProperties(fname=fpath)

# set title, lables, text with the font
ax.set_title("Custom font title", fontproperties=prop, size=20)
ax.set_xlabel("X label", fontproperties=prop, size=20)
ax.set_ylabel("Y label", fontproperties=prop, size=20)
plt.text(0.2, 0.5, "random text", fontproperties=prop, size=20)

# set axis tick labels with the font
for tick in ax.get_xticklabels():
    tick.set_fontproperties(prop)
    
for tick in ax.get_yticklabels():
    tick.set_fontproperties(prop)
```
